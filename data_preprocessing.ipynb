```python
duplicate_rows = df.duplicated()
num_duplicates = duplicate_rows.sum()

print("Number of duplicate rows found:", num_duplicates)
```

```python
df = df.drop_duplicates()
```

```python
expense_cols = [
    'rent',
    'groceries',
    'transport',
    'utilities',
    'dining_out',
    'shopping',
    'entertainment',
    'healthcare',
    'education',
    'total_emi'
]
```

```python
# Replace text-based missing values with NaN
df[expense_cols] = df[expense_cols].replace(
    ['NA', 'N/A', 'na', 'n/a', '', ' '],
    np.nan
)

# Treat zero as missing
df[expense_cols] = df[expense_cols].replace(0, np.nan)
```

```python
print("Missing values before imputation:")
print(df[expense_cols].isna().sum())
```

```python
# Apply user-wise median imputation
for col in expense_cols:
    df[col] = df.groupby('user_id')[col].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
print("Missing values after imputation:")
print(df[expense_cols].isna().sum())

print("\nZero values after imputation:")
print((df[expense_cols] == 0).sum())
```

```python
expense_cols = [
    'rent',
    'groceries',
    'transport',
    'utilities',
    'dining_out',
    'shopping',
    'entertainment',
    'healthcare',
    'education',
    'total_emi'
]
```

```python
# Recalculate total expense as sum of all expense features
df['total_expense'] = df[expense_cols].sum(axis=1)
```

```python
# Expense ratio = total_expense / monthly_income
df['expense_ratio'] = df['total_expense'] / df['monthly_income']
```

```python
# Net balance = income - total_expense
df['net_balance'] = df['monthly_income'] - df['total_expense']
```

```python
# Savings rate = net_balance / monthly_income
df['savings_rate'] = df['net_balance'] / df['monthly_income']
```

```python
# Check for invalid values
print("Any NaN in derived features:")
print(df[['total_expense', 'expense_ratio', 'net_balance', 'savings_rate']].isna().sum())

print("\nSummary statistics of derived features:")
df[['total_expense', 'expense_ratio', 'net_balance', 'savings_rate']].describe()
```

```python
def spending_behaviour(row):
    if row['expense_ratio'] <= 0.7:
        return 'Low'
    elif row['expense_ratio'] <= 1.0:
        return 'Moderate'
    else:
        return 'High'

df['spending_behaviour'] = df.apply(spending_behaviour, axis=1)
```

```python
def savings_behaviour(row):
    if row['savings_rate'] >= 0.2:
        return 'Good'
    elif row['savings_rate'] >= 0:
        return 'Average'
    else:
        return 'Poor'

df['savings_behaviour'] = df.apply(savings_behaviour, axis=1)
```

```python
# Recalculate EMI ratio
df['emi_ratio'] = df['total_emi'] / df['monthly_income']

def emi_status(row):
    if row['emi_ratio'] <= 0.3:
        return 'Safe'
    else:
        return 'Risky'

df['emi_status'] = df.apply(emi_status, axis=1)
```

```python
def persona(row):
    if row['savings_behaviour'] == 'Good' and row['spending_behaviour'] in ['Low', 'Moderate']:
        return 'Saver'
    elif row['spending_behaviour'] == 'High' and row['savings_behaviour'] == 'Poor':
        return 'Spender'
    else:
        return 'Balanced'

df['persona'] = df.apply(persona, axis=1)
```

```python
print("Spending Behaviour Distribution:")
print(df['spending_behaviour'].value_counts())

print("\nSavings Behaviour Distribution:")
print(df['savings_behaviour'].value_counts())

print("\nEMI Status Distribution:")
print(df['emi_status'].value_counts())

print("\nPersona Distribution:")
print(df['persona'].value_counts())
```
